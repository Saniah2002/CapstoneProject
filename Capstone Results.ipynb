# Popularity Based Model

```python
import pandas as pd
import ast 

class ModifiedPopularityBasedRecommender:
    def __init__(self, items_df):
        self.items_df = items_df
        # Assume 'purchase_count' is calculated as before and added to 'items_df'
        self.popular_products = self._get_popular_products()
        
    def _get_popular_products(self):
        # Count the occurrences of each product_id in items_df to find the popularity
        product_counts = self.items_df['product_id'].value_counts().reset_index()
        product_counts.columns = ['product_id', 'purchase_count']
        # Merge with items_df to get product details, dropping duplicates
        popular_products_df = product_counts.merge(self.items_df.drop_duplicates(subset='product_id'), on='product_id')
        return popular_products_df

    def get_similar_popular_products(self, product_id, top_n=10):
        # Find the category of the given product_id
        product_category = self.items_df.loc[self.items_df['product_id'] == product_id, 'category'].values[0]
        
    
        # Filter the popular_products DataFrame for items in the same category
        similar_products = self.popular_products[self.popular_products['category'] == product_category]
        
        # Sort by 'purchase_count' to get the most popular items
        recommendations = similar_products.sort_values(by='purchase_count', ascending=False)
        
        # Exclude the given product_id from the recommendations
        recommendations = recommendations[recommendations['product_id'] != product_id]
        
        return recommendations.head(top_n)

# Example usage

recommender = ModifiedPopularityBasedRecommender(items_df)
product_id = 33927 # some product_id from your dataset
top_n = 100
similar_popular_products = recommender.get_similar_popular_products(product_id, top_n)
similar_popular_products
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Assuming items_df is already available

# Step 1: Create a target variable for popularity within category

# Define popular as being in the top 25% of purchase counts within its category
thresholds = similar_popular_products.groupby('category')['purchase_count'].quantile(0.75).reset_index()
thresholds.columns = ['category', 'popularity_threshold']
similar_popular_products = similar_popular_products.merge(thresholds, on='category', how='left')
similar_popular_products['is_popular'] = (similar_popular_products['purchase_count'] >= similar_popular_products['popularity_threshold']).astype(int)

# Step 2: Prepare your features (X) and target variable (y)
# Let's include 'category', 'brand', and 'type' as features, assuming they're relevant
X = similar_popular_products[['sub_category', 'brand','type']]
X = pd.get_dummies(X, drop_first=True)  # One-hot encode categorical variables

# Including numerical features as well, assuming 'sale_price' and 'market_price' are relevant
numerical_features = similar_popular_products[['sale_price', 'market_price','rating']]
X = pd.concat([X, numerical_features], axis=1)

y = similar_popular_products['is_popular']

# Step 3: Split the data, train the model, and evaluate
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, stratify=y, random_state=42)

# Use a pipeline to scale numerical features
pipeline = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))
pipeline.fit(X_train, y_train)

# Predict and evaluate
predictions = pipeline.predict(X_test)
print(classification_report(y_test, predictions))
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

class CategoryBasedRecommender:
    def __init__(self, items_df):
        self.items_df = items_df.set_index('product_id')  # Set the index to product_id
        self.items_df['product'] = self.items_df['product'].astype(str)
        self.item_profiles = self._create_item_profiles()
        self.item_similarity_matrix = self._calculate_similarity()

    def _create_item_profiles(self):
        count_vectorizer = CountVectorizer()
        count_matrix = count_vectorizer.fit_transform(self.items_df['product'])
        item_profiles = pd.DataFrame(count_matrix.toarray(), index=self.items_df.index,
                                     columns=count_vectorizer.get_feature_names_out())
        return item_profiles

    def _calculate_similarity(self):
        cosine_sim_matrix = pd.DataFrame(cosine_similarity(self.item_profiles),
                                         index=self.item_profiles.index,
                                         columns=self.item_profiles.index)
        return cosine_sim_matrix

    def get_recommendations(self, product_id, top_n=10):
        # Ensure product_id is an integer for proper row slicing
        product_id = int(product_id)
        
        # Check if product_id exists in the index
        if product_id not in self.item_similarity_matrix.index:
            raise KeyError(f"Product ID {product_id} not found in the item similarity matrix.")
        
        # Get the similarity scores for the item
        item_similarities = self.item_similarity_matrix.loc[product_id]
        
        # Ensure item_similarities is a Series
        if isinstance(item_similarities, pd.DataFrame):
            item_similarities = item_similarities.squeeze()  # Convert to Series if it's a DataFrame
        
        # Sort the items based on the similarity scores
        similar_items = item_similarities.sort_values(ascending=False)
        
        # Exclude the item itself from the recommendation
        similar_items = similar_items[similar_items.index != product_id]
        
        # Get the top N similar items
        top_similar_items = similar_items.head(top_n).index
        
        # Return the full details of the recommended items
        return self.items_df.loc[top_similar_items]


# Instantiate the recommender
recommender = CategoryBasedRecommender(items_df)

# Get recommendations for a specific product ID
product_id = 33927  # Use an actual product ID from your dataset
top_n = 100  # The number of top recommendations you want

recommendations = recommender.get_recommendations(product_id, top_n)
recommendations
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

class CollaborativeFilteringRecommender:
    def __init__(self, items_df):
        self.items_df = items_df
        # Create a user-item rating matrix
        self.user_item_matrix = self._create_interaction_matrix()

    def _create_interaction_matrix(self):
        # Group by user_id and product_id and aggregate ratings
        # Assuming 'rating' column exists in your interactions_df
        interaction_matrix = self.items_df.pivot_table(index='user_id', columns='product_id', values='rating').fillna(0)
        return interaction_matrix

    def get_recommendations(self, user_id, items_df, top_n=10):
        # Compute cosine similarity between all users
        user_similarity_matrix = cosine_similarity(self.user_item_matrix)
        user_index = self.user_item_matrix.index.tolist().index(user_id)

        # Get similarity scores for the target user and sort them
        user_similarity_scores = user_similarity_matrix[user_index]
        similar_users_indices = user_similarity_scores.argsort()[-top_n-1:-1][::-1]

        # Aggregate the products rated by similar users
        recommended_products = set()
        for idx in similar_users_indices:
            similar_user_id = self.user_item_matrix.index[idx]
            products_rated_by_similar = self.user_item_matrix.loc[similar_user_id]
            # Consider products with a rating above a certain threshold, e.g., more than 0
            high_rated_products = products_rated_by_similar[products_rated_by_similar > 0].index.tolist()
            recommended_products.update(high_rated_products)

        # Remove the products already rated by the target user
        products_rated_by_user = set(self.user_item_matrix.loc[user_id][self.user_item_matrix.loc[user_id] > 0].index)
        recommended_products.difference_update(products_rated_by_user)

        # Return a DataFrame with details for each recommended product ID
        recommended_products_list = list(recommended_products)[:top_n]
        recommended_products_details = items_df.loc[items_df['product_id'].isin(recommended_products_list)]

        return recommended_products_details

```

# Hybrid Model

```python
import pandas as pd

class HybridRecommender:
    def __init__(self, content_model, collaborative_model, popularity_model, items_df, weights):
        self.content_model = content_model
        self.collaborative_model = collaborative_model
        self.popularity_model = popularity_model
        self.items_df = items_df
        self.weights = weights


    def get_recommendations(self, product_id, user_id, top_n=10):
        # Get recommendations from each model
        content_recommendations = self.content_model.get_recommendations(product_id, top_n)
        collaborative_recommendations = self.collaborative_model.get_recommendations(user_id, self.items_df, top_n)
        popularity_recommendations = self.popularity_model.get_similar_popular_products(product_id, top_n)

        # Create a DataFrame to aggregate scores
        recommendations_df = pd.DataFrame(index=self.items_df.index)

        # Aggregate scores from each model based on their indices and assigned weights
        recommendations_df['content_score'] = recommendations_df.index.isin(content_recommendations).astype(int) * self.weights['content']
        recommendations_df['collaborative_score'] = recommendations_df.index.isin(collaborative_recommendations).astype(int) * self.weights['collaborative']
        recommendations_df['popularity_score'] = recommendations_df.index.isin(popularity_recommendations).astype(int) * self.weights['popularity']

        # Sum up the weighted scores to a total score column
        recommendations_df['total_score'] = recommendations_df[['content_score', 'collaborative_score', 'popularity_score']].sum(axis=1)

        # Sort by total score and retrieve the top N recommendations
        top_recommendations = recommendations_df.nlargest(top_n, 'total_score').index
        return self.items_df.loc[top_recommendations]

# Define your models
content_model = CategoryBasedRecommender(items_df)
collaborative_model = CollaborativeFilteringRecommender(items_df)
popularity_model = ModifiedPopularityBasedRecommender(items_df)

# Define weights for each model based on effectiveness or domain knowledge
weights = {'content': 0.4, 'collaborative': 0.4, 'popularity': 0.2}

# Create the Hybrid Recommender
hybrid_recommender = HybridRecommender(content_model, collaborative_model, popularity_model, items_df, weights)

# Get hybrid recommendations
hybrid_recommendations = hybrid_recommender.get_recommendations(product_id=33927, user_id=2746, top_n=100)
hybrid_recommendations
```
